### TASK 2.1 TRACK MULTIPLICITIES 

```python
import glob
import csv

directory = "./Vertices"
csv_files = glob.glob(f"{directory}/*.csv") 

multiplicities = []

for file_ in csv_files:
    with open(file_, mode ='r') as file:    
       csvFile = csv.DictReader(file)
       for lines in csvFile:
            multiplicities.append(int(lines['mult']))
```

```python
print(len(multiplicities))
```

```python
import os

directory = "./"  
file_path = os.path.join(directory, "multiplicities.txt")

# Writing multiplicities to file
with open(file_path, 'w') as f:
    for i, m in enumerate(multiplicities):
        f.write(str(m))
        if i < len(multiplicities) - 1:  
            f.write('\n')
```

```python
print(min(multiplicities)," ",max(multiplicities)," ",len(multiplicities))
```

### TASK 2.2 ANGLES OF MUON TRACKS 

```python
import math
directory = "./Tracks"
csv_files = glob.glob(f"{directory}/*.csv") 

Angles_XZ = []
Angles_YZ = []
count = 0

for file_ in csv_files:
    with open(file_, mode ='r') as file:    
       csvFile = csv.DictReader(file)
       for lines in csvFile:
            if(int(lines['trType'])) == 1:
                angle_xz = math.atan(float(lines['slopeXZ']))*(180/math.pi)
                Angles_XZ.append(angle_xz)
                angle_yz = math.atan(float(lines['slopeYZ']))*(180/math.pi)
                Angles_YZ.append(angle_yz)
```

```python
print(len(Angles_XZ),len(Angles_YZ))
```

```python
directory = "./" 
file_path = os.path.join(directory, "Angles_XZ.txt")

# Writing the list to the file
with open(file_path, 'w') as f:
    for i, angle in enumerate(Angles_XZ):
        f.write(str(angle))
        if i < len(Angles_XZ) - 1:  
            f.write('\n') 

print(f"File saved at: {file_path}")
```

```python
directory = "./"  
file_path = os.path.join(directory, "Angles_YZ.txt")

with open(file_path, 'w') as f:
    for i, angle in enumerate(Angles_YZ):
        f.write(str(angle))
        if i < len(Angles_YZ) - 1:  
            f.write('\n')
```
